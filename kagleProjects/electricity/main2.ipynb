# Interpolation task


```python
#including important libraries
import numpy as np
import pandas as pd
from scipy.interpolate import interp1d
import matplotlib.pyplot as plt
from matplotlib.pyplot import rcParams
rcParams['figure.figsize'] = 16,8
```

```python
data = pd.read_csv("train.0.0.csv")
data.timestamp = pd.to_datetime(data.timestamp)
data['hour'] = data['timestamp'].dt.hour
```

```python
data_hours = data.groupby("hour") #створює обєкт з сорованими по часі даними
data_avg = pd.DataFrame({
    "Avg": data_hours["meter_reading"].mean(),
    "Median":data_hours["meter_reading"].median()
})

data_avg.plot()
plt.show()
```

```python
#median and average has large diference. Firstly you should to decrise this diff 
filtred_data_hours = data[data["meter_reading"]>0].groupby("hour")

clear_data_avg = pd.DataFrame({
    "Avg": filtred_data_hours["meter_reading"].mean(),
    "Median":filtred_data_hours["meter_reading"].median()
})

clear_data_avg.plot()
plt.show()
```

```python
x = np.arange(0,24) #create array with hours 0 to 23
y = interp1d(x, filtred_data_hours["meter_reading"].median(), kind="cubic")
xn = np.arange(0,23.1,0.1)
yn = y(xn)

plt.plot(x,  filtred_data_hours["meter_reading"].median(), 'x', xn, yn, '-')
```
