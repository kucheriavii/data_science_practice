# Як зробити афігєнну таблицю кореляцій

```python
import numpy as np #імпорт бібліотеки
masc_corr = np.zeros_like(data_corr) #створити нульову матрицю 
triangle_indexes = np.triu_indices_from(masc_corr) #вибрати індекси вище головної діагоналі
masc_corr[triangle_indexes] = 1 #присвоїти цим індексам значення 1
masc_corr #пощупати очима
```

```python
plt.figure(figsize=(10,5))
sns.heatmap(data.corr(), mask=masc_corr, annot=True, annot_kws={"size":10}, fmt='.2f') #намалювати таблицю
plt.xticks(fontsize=11)
plt.yticks(fontsize=11)
plt.show()
```

```python
data.corr()
```

```python
nox_dis_corr = round(data["nox"].corr(data['dis']),3)

plt.figure(figsize=(9,6))
plt.title(f"DIS vs NOX (correlation {nox_dis_corr})", fontsize=14)
plt.scatter(data['dis'], data['nox'], alpha=0.6, s=80, color="indigo")

plt.xlabel('dis - distance from employment', fontsize=12)
plt.ylabel('nox - nitric oxide pollution', fontsize=12)
plt.show()
```

```python
sns.set()
sns.set_context('notebook')
sns.set_style('whitegrid')
sns.jointplot(x=data['dis'], y=data['nox'], size=7, color='indigo', joint_kws={'alpha':0.5})
plt.show()
```

```python
sns.set()
sns.set_context('notebook')
sns.set_style('whitegrid')
sns.jointplot(x=data['dis'], y=data['nox'], kind="hex", color='blue')
plt.show()
```
